# Preprocessing Data

Daqui eu vou querer o PSxG por chute tentado (PSxG/SoT)

```python
gk_advanced_stats_df.head()
```

Daqui eu quero quantidade de defesas e o Min

```python
gk_basic_individual_stats_df.head()
```

```python
gk_advanced_stats_df = gk_advanced_stats_df[['Rk', 'Player', 'PSxG/SoT']]
gk_basic_individual_stats_df = gk_basic_individual_stats_df[['Rk', 'Player', 'Saves', 'Min']]
```

```python
gk_advanced_stats_df.head()
```

```python
gk_basic_individual_stats_df.head()
```

```python
gk_stats_df = pd.merge(gk_advanced_stats_df, gk_basic_individual_stats_df, how = 'inner', on = 'Rk')
assert gk_stats_df['Player_x'].equals(gk_stats_df["Player_y"])
gk_stats_df.drop(columns=['Player_y'], inplace=True)
```

```python
gk_stats_df['Player'] = gk_stats_df['Player_x']
gk_stats_df.drop(columns=['Player_x'], inplace=True)
```

```python
gk_stats_df = gk_stats_df[['Rk', 'Player', 'Min', 'Saves', 'PSxG/SoT', ]]
```

```python
gk_stats_df.head()
```

```python
gk_stats_df['PSxG/SoT'] = gk_stats_df['PSxG/SoT'].astype(float)
gk_stats_df['Saves'] = gk_stats_df['Saves'].astype(int)
```

# Analysis

```python
X = gk_stats_df['Saves'].to_numpy(dtype=int)
Y = gk_stats_df['PSxG/SoT'].to_numpy(dtype=float)
names = gk_stats_df['Player']
minutes_played = [0.05 * min_played for min_played in gk_stats_df['Min'].to_numpy(dtype=float)]
```

## Light Mode

```python
plt.rcParams.update({'font.family':'Ubuntu'})
bgcol = "#efefef"

fig, ax = plt.subplots(figsize=(10, 7), dpi=120)
fig.set_facecolor(bgcol)
ax.set_facecolor(bgcol)
ax.scatter(X, Y, s=minutes_played, alpha=0.6, edgecolors="black", linewidths=0.5)

for i, nome in enumerate(names):
    ax.text(
        X[i], # Pequeno deslocamento no eixo X para não ficar em cima do ponto
        Y[i] - 0.006,       # Posição Y
        nome,
        fontsize=7, # Tamanho da fonte pequeno para não poluir muito
        color='#333333',
        ha='center'
    )

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_color('#ccc8c8')
ax.spines['bottom'].set_color('#ccc8c8')

plt.tick_params(axis='x', labelsize=12, color='#ccc8c8')
plt.tick_params(axis='y', labelsize=12, color='#ccc8c8')

mean_x = X.mean()
mean_y = Y.mean()

plt.hlines(mean_y, X.min(), X.max(), color='#c2c1c0')
fig.text(.05,.14,'Quantidade de Defesas', size=9, color='#575654',rotation=90)

plt.vlines(mean_x, Y.min(), Y.max(), color='#c2c1c0')
fig.text(.12,0.05,'PSxG por quantidade de chutes tomados', size=9, color='#575654')

fig.text(.32,.98,'Defesas vs PSxG/SoT',size=20)
fig.text(.15,.93,'Comparação da quantidade de defesas feitas com a qualidade média dessas defesas', size=12)

fig.text(.85,.495,'Média de Defesas', size=6, color='#575654')
fig.text(.485, .17,'Média de PSxG', size=6, color='#575654',rotation=90)

style_box = dict(boxstyle='round,pad=0.4', fc='white', ec='#aaaaaa', lw=1)

ax.text(X.min() - 5, mean_y + 0.08, 'Poucas Defesas\nPorém no geral difíceis',
        ha='left', va='bottom', fontsize=8, color='#333333', bbox=style_box)

ax.text(mean_x + 50, mean_y + 0.08, 'Muitas Defesas\nE geralmente difíceis',
        ha='left', va='bottom', fontsize=8, color='#333333', bbox=style_box)

ax.text(X.min() + 32, mean_y - 0.05, 'Poucas Defesas\nE geralmente fáceis',
        ha='left', va='top', fontsize=8, color='#333333', bbox=style_box)

ax.text(mean_x + 50, mean_y - 0.05, 'Muitas Defesas\nPorém não tão difíceis',
        ha='left', va='top', fontsize=8, color='#333333', bbox=style_box)

plt.tight_layout(rect=[0.05, 0.05, 0.98, 0.90]) # Ajusta o layout para os textos não cortarem
plt.show();
```
