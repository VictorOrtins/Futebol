```python
strings[:100]
```

### Shots

#### Decoding Data

```python
str_start = strings.index("('") + 2
str_end = strings.index("')")
json_data = strings[str_start:str_end]
json_data = json_data.encode('utf-8').decode('unicode_escape')

data = json.loads(json_data)
```

```python
data
```

```python
data['h']
```

```python
data['a']
```

#### Converting to Dataframe

```python
data_home = pd.json_normalize(data['h'])
data_away = pd.json_normalize(data['a'])

df_home = pd.DataFrame(data_home)
df_away = pd.DataFrame(data_away)

df = pd.concat([df_home, df_away])
df['minute'] = df['minute'].astype(int)
df.sort_values(by=['minute'], inplace=True)

df.head()
```

## Dates Data

```python
dates_string = scripts[1].string
```

```python
str_start = dates_string.index("('") + 2
str_end = dates_string.index("')")
json_data = dates_string[str_start:str_end]
json_data = json_data.encode('utf-8').decode('unicode_escape')

data = json.loads(json_data)
```

```python
data
```

```python
games_df = pd.DataFrame(data)
```

```python
games_df.head()
```

```python
games_df['forecast'].iloc[0]
```

```python
games_df['home_team_id'] = games_df['h'].apply(lambda x: x['id'])
games_df['home_team_name'] = games_df['h'].apply(lambda x: x['title'])
games_df['home_team_short_name'] = games_df['h'].apply(lambda x: x['short_title'])

games_df['away_team_id'] = games_df['a'].apply(lambda x: x['id'])
games_df['away_team_name'] = games_df['a'].apply(lambda x: x['title'])
games_df['away_team_short_name'] = games_df['a'].apply(lambda x: x['short_title'])

games_df['home_goals'] = games_df['goals'].apply(lambda x: x['h'])
games_df['away_goals'] = games_df['goals'].apply(lambda x: x['a'])

games_df['home_xG'] = games_df['xG'].apply(lambda x: x['h'])
games_df['away_xG'] = games_df['xG'].apply(lambda x: x['a'])

games_df['forecast_win'] = games_df['forecast'].apply(lambda x: x['w'])
games_df['forecast_draw'] = games_df['forecast'].apply(lambda x: x['d'])
games_df['forecast_loss'] = games_df['forecast'].apply(lambda x: x['l'])
```

```python
games_df.head()
```

```python
games_df.drop(columns=['h', 'a', 'goals', 'xG'], inplace=True)
```

```python
games_df.head()
```

## Player Stats

```python
str_start = player_stats_string.index("('") + 2
str_end = player_stats_string.index("')")
json_data = player_stats_string[str_start:str_end]
json_data = json_data.encode('utf-8').decode('unicode_escape')

data = json.loads(json_data)
```

```python
data
```

```python
data.keys()
```

```python
pd.DataFrame(data)
```

```python
data['AMC']
```

```python
new_data = {}
for position, position_dict in data.items():
    if position not in new_data:
        new_data[position] = {}
    for stat, stats in position_dict.items():
        for math, stat_value in stats.items():
            new_data[f'{position}'][f'{stat}_{math}'] = stat_value
```
